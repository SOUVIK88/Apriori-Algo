```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
pd.set_option('display.max_rows',10)
pd.set_option('display.max_columns',10)
```

```python
df=pd.read_csv('OnlineRetail.csv',encoding='latin1')
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
df['InvoiceNo']=df['InvoiceNo'].astype(str)
df=df[~df['InvoiceNo'].str.contains('C')]
df.shape
```

```python
basket=(df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index(['InvoiceNo']))
```

```python
basket.head()
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_set=basket.applymap(encode_units)
basket_set.head()
```

```python
frequent_items=apriori(basket_set,min_support=0.07,use_colnames=True)
rules=association_rules(frequent_items,metric='lift',min_threshold=1)
rules.head()
```

```python
rules[(rules['lift']>=6) & (rules['confidence']>=0.8)]
```
